#### Read all SraRunTables into meta_table

```python
files = [file for file in listdir() if ("SraRunTable" in file and file.endswith(".txt"))]
```

```python
meta_table = pd.read_csv(files[0])
geo_id = files[0].split("_")[1][:-4]
meta_table["geo_id"] = geo_id
for file in files[1:]:
    table = pd.read_csv(file)
    geo_id = file.split("_")[1][:-4]
    table["geo_id"] = geo_id
    meta_table = meta_table.append(table) 
```

```python
inputs = meta_table.loc[meta_table["antibody"] == "input"]
def get_sra_id_of_control(row, inputs):
    if row["treatment"] in inputs[(inputs["source_name"] == row["source_name"]) & (inputs["SRA Study"] == row["SRA Study"])]["treatment"].unique():
        return inputs[(inputs["source_name"] == row["source_name"]) & (inputs["SRA Study"] == row["SRA Study"]) & (inputs["treatment"] == row["treatment"])]["Run"].to_list()
    else:
        return inputs[(inputs["source_name"] == row["source_name"]) & (inputs["SRA Study"] == row["SRA Study"])]["Run"].to_list()
meta_table["controls"] = meta_table.apply(get_sra_id_of_control, inputs=inputs, axis=1)
```

```python
broad_marks = ["H3F3A", "H3K27me3", "H3K36me3", "H3K4me1", "H3K79me2", "H3K79me3", "H3K9me1", "H3K9me2", "H4K20me1"]
narrow_marks = ["H2AFZ", "H3ac", "H3K27ac", "H3K4me2", "H3K4me3", "H3K9ac"]
exceptions = ["H3K9me3"]
```

```python
ab = meta_table["antibody"].unique()
for i in ab:
    if i in broad_marks:
        print(i, "broad_mark")
    if i in narrow_marks:
        print(i, "narrow_mark")
    if i in exceptions:
        print(i, "exception")
```

```python
def get_peak_width(row):
    target = row["antibody"]
    if target in broad_marks:
        return "broad"
    elif target in narrow_marks:
        return "narrow"
    elif target in exceptions:
        return "exception"
    else:
        return np.nan
        
meta_table["peak_width"] = meta_table.apply(get_peak_width, axis=1)
```

```python
healthy = ["Healthy", "healthy", "Young"]
diseases = [label for label in meta_table["disease_state"].unique() if label not in healthy and pd.notna(label)]
```

```python
def get_binary_disease_state(x):
    if x in diseases:
        return 1
    elif x in healthy:
        return 0
    else:
        return np.nan
```

```python
meta_table["has_disease"] = meta_table["disease_state"].apply(get_binary_disease_state)
```
