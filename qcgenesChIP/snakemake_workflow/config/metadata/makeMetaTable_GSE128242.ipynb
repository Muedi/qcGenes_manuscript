```python
import pandas as pd
import re
import os
```

```python
path = "raw/GSE128242_SraRunTable.txt"
df = pd.read_csv(path)
```

```python
geo_id = os.path.split(path)[-1].split("_")[0]
```

```python
df["geo_id"] = geo_id
```

```python
df["is_sample"] = df["chip_antibody"].apply(lambda x : 0 if x == "none" else 1)
```

```python
df["is_healthy"] = df["Genotype"].apply(lambda x : 0 if x == "AP-1 depleted" else 1)
```

```python
df["condition"] = df["is_healthy"].apply(lambda x : "AP-1 depleted leomyomas cells" if x == 0 else "leomyomas cells")
```
