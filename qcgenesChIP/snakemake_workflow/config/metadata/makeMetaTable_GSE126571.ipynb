```python
import pandas as pd
import re
import os
```

```python
path = "raw/GSE126571_SraRunTable.txt"
df = pd.read_csv(path)
```

```python
geo_id = os.path.split(path)[-1].split("_")[0]
```

```python
df["geo_id"] = geo_id
```

```python
df = df.loc[df["tissue"] != "heart", :]
```

```python
df["is_sample"] = df["Antibody"].apply(lambda x : 0 if x == "-" else 1)
```

```python
df["is_healthy"] = df["disease"].apply(lambda x : 1 if x == "-" else 0)
```

```python
df["condition"] = df["disease"].apply(lambda x : "healthy tissue" if x == "-" else "cardiomyopathy tissue")
```

```python
# Input Information completely omitted in this dataset
# df["is_sample"] = df["chip_antibody"].apply(lambda x : 0 if re.search(r"input", x) else 1)
```

```python
df.to_csv("GSE126571_metaTable.csv")
```
