```python
import pandas as pd
import re
import os
```

```python
path = "raw/GSE104404_SraRunTable.txt"
df = pd.read_csv(path)
```

```python
df.iloc[:,9:19]
```

```python
geo_id = os.path.split(path)[-1].split("_")[0]
```

```python
df["geo_id"] = geo_id
```

```python
df["is_sample"] = df["chip_antibody"].apply(lambda x : 0 if x == "none" else 1)
```

```python
df["donor_age"] = df["donor_age"].apply(lambda x : 29 if x == "<30" else x).astype(int)
```

```python
df["is_healthy"] = df["donor_age"].apply(lambda x : 0 if x > 50 else 1)
```

```python
df["condition"] = df["donor_age"].apply(lambda x : "old donor HSCs" if x > 50 else "young donor HSCs")
```

```python
df.to_csv("GSE104404_metaTable.csv")
```
