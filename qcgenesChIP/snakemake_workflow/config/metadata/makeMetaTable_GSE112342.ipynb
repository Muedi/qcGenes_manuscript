```python
import pandas as pd
import re
import os
```

```python
path = "raw/GSE112342_SraRunTable.txt"
df = pd.read_csv(path)
```

```python
df.columns
```

```python
geo_id = os.path.split(path)[-1].split("_")[0]
```

```python
df["geo_id"] = geo_id
```

```python
df["is_sample"] = df["chip_antibody"].apply(lambda x : 0 if x == "None" else 1)
```

```python
df.disease_state.value_counts()
```

```python
df["is_healthy"] = df["disease_state"].apply(lambda x : 1 if x == "control" else 0)
```

```python
df["condition"] = df["disease_state"].apply(lambda x : "healthy T cells" if x == "control" else "Type 1 diabetes T cells")
```
